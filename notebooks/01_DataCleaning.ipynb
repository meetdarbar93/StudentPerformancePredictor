# 01 - Data Cleaning
Load raw_dataset.csv, clean, and save cleaned_data.csv

```python
import pandas as pd

# Load dataset
df = pd.read_csv("../data/raw_dataset.csv")

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Participation'] = le.fit_transform(df['Participation'])
```

```python
# Filling missing values based on column type

df['Hours_Studied'].fillna(df['Hours_Studied'].mean())
df['Attendance'].fillna(df['Attendance'].mean())
df['Previous_Score'].fillna(df['Previous_Score'].median())
df['Assignments_Completed'].fillna(df['Assignments_Completed'].median())
df['Sleep_Hours'].fillna(df['Sleep_Hours'].mean())

df['Participation'].fillna(df['Participation'].mode()[0])

df['Pass_Fail'].fillna(df['Pass_Fail'].mode()[0])

df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)

df.shape
```

```python
df.to_csv("../data/cleaned_data.csv", index=False)
print("Cleaned dataset saved!")
```
